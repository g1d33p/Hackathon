```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = "/Users/jeevandeep/Downloads/Hackathon/Notebooks/combined_medium_high_veryhigh_claims.csv"  # Update the path to your local file
data_df = pd.read_csv(data_path)

# Drop duplicates to ensure clean data
data_df = data_df.drop_duplicates()

# Sorting the data based on priority levels (Medium, High, Very High) in descending order
priority_order = ['Medium', 'High', 'Very High']
data_df['Scale Claim Cost'] = pd.Categorical(data_df['Scale Claim Cost'], categories=priority_order, ordered=True)
sorted_data = data_df.sort_values(by='Scale Claim Cost', ascending=False)

# Visualize Style counts in descending order of priority levels
plt.figure(figsize=(12, 8))
style_bar = sns.countplot(data=sorted_data, x='Style', order=sorted_data['Style'].value_counts().index, palette="viridis")
plt.title('Styles in Descending Order of Priority Levels')
plt.xlabel('Style')
plt.ylabel('Count of Trucks')
plt.xticks(rotation=45)
plt.tight_layout()

# Add counts above bars
for p in style_bar.patches:
    style_bar.annotate(format(p.get_height(), '.1f'), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points')

plt.show()

# Visualizing Attributes and Option Codes
attribute_columns = [col for col in sorted_data.columns if "Attribute" in col]

# Generating countplots for each attribute
for attribute in attribute_columns:
    plt.figure(figsize=(12, 6))
    attribute_bar = sns.countplot(data=sorted_data, x=attribute, order=sorted_data[attribute].value_counts().index, palette="coolwarm")
    plt.title(f'Options for {attribute} in Descending Priority')
    plt.xlabel(f'{attribute} Options')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Add counts above bars
    for p in attribute_bar.patches:
        attribute_bar.annotate(format(p.get_height(), '.1f'), 
                               (p.get_x() + p.get_width() / 2., p.get_height()), 
                               ha = 'center', va = 'center', 
                               xytext = (0, 9), 
                               textcoords = 'offset points')
    
    plt.show()
```
